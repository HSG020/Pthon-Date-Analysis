```python
# 获取输入字符串
s = input("请输入一个字符串：")

# 初始化计数器
uppercase_count = 0
lowercase_count = 0
digit_count = 0

# 遍历字符串
for char in s:
    if char.isupper():
        uppercase_count += 1
    elif char.islower():
        lowercase_count += 1
    elif char.isdigit():
        digit_count += 1

# 输出结果
print(f"大写字母个数: {uppercase_count}")
print(f"小写字母个数: {lowercase_count}")
print(f"数字个数: {digit_count}")
```

```python
def extract_birthdate(id_number):
    """
    从身份证号码中提取出生年月日。

    参数:
        id_number (str): 身份证号码，长度应为18位或15位。

    返回:
        str: 出生日期，格式为 YYYY-MM-DD。
    """
    if len(id_number) == 18:
        birthdate_str = id_number[6:14]  # 第7到14位
        year = birthdate_str[0:4]
        month = birthdate_str[4:6]
        day = birthdate_str[6:8]
    elif len(id_number) == 15:
        birthdate_str = id_number[6:12]  # 第7到12位
        year = "19" + birthdate_str[0:2]
        month = birthdate_str[2:4]
        day = birthdate_str[4:6]
    else:
        return "无效的身份证号码长度。"

    # 验证日期的有效性
    try:
        from datetime import datetime
        birthdate = datetime(int(year), int(month), int(day))
        return birthdate.strftime("%Y-%m-%d")
    except ValueError:
        return "身份证号码中的出生日期无效。"

# 示例用法
if __name__ == "__main__":
    id_num = input("请输入身份证号码：").strip()
    birthdate = extract_birthdate(id_num)
    print(f"出生日期：{birthdate}")
```

```python
import random
import re

# 定义姓氏字符串，使用正则表达式去除所有非汉字字符
first = '''赵钱孙李，周吴郑王。冯陈褚卫，蒋沈韩杨。朱秦尤许，何吕施张。'''
# 提取所有汉字字符作为姓氏列表
surnames = re.findall(r'[\u4e00-\u9fff]', first)

# 定义名字字符集
# 这里提供了一些常用的名字字符，你可以根据需要自行添加
second = '''伟刚勇毅俊峰强军平保东文辉力明永健世广志义兴良海山'''
# 将名字字符转换为列表
given_names = list(second)

def generate_random_name():
    """
    生成一个随机的三字中文姓名。
    
    返回:
        str: 生成的姓名
    """
    # 随机选择一个姓氏
    surname = random.choice(surnames)
    
    # 随机选择两个名字字符
    given_name = ''.join(random.choices(given_names, k=2))
    
    # 组合成完整的姓名
    full_name = surname + given_name
    return full_name

# 示例输出
if __name__ == "__main__":
    random_name = generate_random_name()
    print(f"随机生成的姓名是：{random_name}")
```

```python
import re

def mask_birthdate(id_number):
    """
    将身份证号码中的出生年月信息替换为星号，并进行基本格式验证。
    
    参数:
        id_number (str): 身份证号码，长度应为15位或18位。
    
    返回:
        str: 处理后的身份证号码，出生年月部分用星号替换，或错误提示。
    """
    id_number = id_number.strip()
    length = len(id_number)
    
    if length == 18:
        pattern = r'^\d{17}[\dXx]$'
        birthdate = id_number[6:14]
    elif length == 15:
        pattern = r'^\d{15}$'
        birthdate = id_number[6:12]
    else:
        return "无效的身份证号码长度。"
    
    if not re.match(pattern, id_number):
        return "身份证号码格式不正确。"
    
    # 简单的出生日期验证
    try:
        if length == 18:
            from datetime import datetime
            datetime.strptime(birthdate, "%Y%m%d")
        elif length == 15:
            from datetime import datetime
            datetime.strptime("19" + birthdate, "%Y%m%d")
    except ValueError:
        return "身份证号码中的出生日期无效。"
    
    # 替换出生年月
    if length == 18:
        masked_id = id_number[:6] + '*' * 8 + id_number[14:]
    else:
        masked_id = id_number[:6] + '*' * 6 + id_number[12:]
    
    return masked_id

# 示例用法
if __name__ == "__main__":
    id_num = input("请输入身份证号码：").strip()
    masked_id = mask_birthdate(id_num)
    print(f"处理后的身份证号码：{masked_id}")
```

```python
# 会员名管理类
class UsernameManager:
    def __init__(self):
        # 使用集合存储标准化后的会员名，集合的查找时间复杂度为O(1)
        self.usernames = set()
    
    def add_username(self, username):
        """
        尝试添加一个新的会员名。

        参数:
            username (str): 用户输入的会员名。

        返回:
            bool: 如果添加成功返回True，否则返回False。
        """
        normalized = self._normalize(username)
        if normalized in self.usernames:
            return False  # 会员名已存在
        self.usernames.add(normalized)
        return True  # 添加成功
    
    def is_username_taken(self, username):
        """
        检查一个会员名是否已经被占用。

        参数:
            username (str): 用户输入的会员名。

        返回:
            bool: 如果已被占用返回True，否则返回False。
        """
        normalized = self._normalize(username)
        return normalized in self.usernames
    
    def _normalize(self, username):
        """
        标准化会员名（例如，转换为全小写）。

        参数:
            username (str): 原始会员名。

        返回:
            str: 标准化后的会员名。
        """
        return username.lower().strip()

# 示例用法
if __name__ == "__main__":
    manager = UsernameManager()
    
    # 预先添加一些会员名
    existing_usernames = ["Alice", "Bob", "CHARLIE", "dave"]
    for name in existing_usernames:
        manager.add_username(name)
    
    while True:
        new_username = input("请输入要注册的会员名（输入'退出'结束）：").strip()
        if new_username.lower() == '退出':
            break
        if manager.is_username_taken(new_username):
            print(f"抱歉，会员名 '{new_username}' 已被占用。请尝试其他名称。")
        else:
            manager.add_username(new_username)
            print(f"会员名 '{new_username}' 注册成功！")
```

```python
import re

def normalize_greeting(s):
    """
    标准化问候语字符串。
    
    参数:
        s (str): 原始字符串。
        
    返回:
        str: 标准化后的字符串。
    """
    # 1. 转换为小写
    s = s.lower()
    
    # 2. 定义替换规则
    replacements = {
        '2': 'to',
        'u': 'you'
    }
    
    # 3. 进行替换
    for key, value in replacements.items():
        # 使用正则确保只替换独立的词
        s = re.sub(r'\b' + re.escape(key) + r'\b', value, s)
    
    # 4. 去除非字母字符（保留空格）
    s = re.sub(r'[^a-z\s]', '', s)
    
    # 5. 去除多余空格
    s = ' '.join(s.split())
    
    return s

# 测试字符串
str1 = '    haPPy BiRthDAy To u'
str2 = 'Happy biRthDAy To u'
str3 = '  haPpy BirThdAy 2 Dear friend'
str4 = '  happy birthday 2 u’'

# 应用标准化函数
normalized_str1 = normalize_greeting(str1)
normalized_str2 = normalize_greeting(str2)
normalized_str3 = normalize_greeting(str3)
normalized_str4 = normalize_greeting(str4)

# 输出结果
print(normalized_str1)  # 输出: happy birthday to you
print(normalized_str2)  # 输出: happy birthday to you
print(normalized_str3)  # 输出: happy birthday to dear friend
print(normalized_str4)  # 输出: happy birthday to you
```
