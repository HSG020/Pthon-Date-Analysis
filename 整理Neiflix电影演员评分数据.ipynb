```python
import ast

# 将 genres 列的字符串转换为列表
df['genres'] = df['genres'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# 使用 explode 将列表拆分成多行
df_exploded = df.explode('genres')

# 按 genres 分组，计算 imdb_score 的最大值
genres_max_scores = df_exploded.groupby("genres")["imdb_score"].max()

# 打印结果
print(genres_max_scores)
```

```python
genres_max_scores_with_person_id = pd.merge(imdb_score_groupby_genres_and_person_id_df, genres_max_scores, on=["genres", "imdb_score"])
genres_max_scores_with_person_id
```
