```python
df = pd.read_csv("owid-covid-data.csv")
```

```python
df.info()
```

```python
df = df.dropna(axis=1, thresh=len(df) * 0.5)
```

```python
df = df.fillna(0)
df.head()
```

```python
us_data = df[df["location"] == "United States"]
```

```python
us_data.loc[:, "date"] = pd.to_datetime(us_data["date"])
```

```python
average_cases = us_data["new_cases"].mean()
average_deaths = us_data["new_deaths"].mean()

print(f"美国每日平均新增确诊: {average_cases}")
print(f"美国每日平均新增死亡: {average_deaths}")
```

```python
us_data.set_index("date",inplace=True)
```

```python
weekly_data = us_data.resample("W").sum()
```

```python
weekly_data.head()
```

```python
import matplotlib.pyplot as plt

# 绘制折线图
plt.figure(figsize=(10, 5))
plt.plot(weekly_data.index, weekly_data['new_cases'], label='Weekly New Cases', color='blue')
plt.plot(weekly_data.index, weekly_data['new_deaths'], label='Weekly New Deaths', color='red')
plt.xlabel('Date')
plt.ylabel('Count')
plt.title('Weekly New COVID-19 Cases and Deaths in the United States')
plt.legend()
plt.show()
```

```python
# 绘制每日新增确诊的柱状图
plt.figure(figsize=(10, 5))
plt.bar(us_data.index, us_data['new_cases'], color='skyblue')
plt.xlabel('Date')
plt.ylabel('New Cases')
plt.title('Daily New COVID-19 Cases in the United States')
plt.show()
```

```python
import seaborn as sns

# 绘制分布图
plt.figure(figsize=(8, 5))
sns.histplot(us_data['new_cases'], kde=True, color='blue')
plt.xlabel('New Cases')
plt.title('Distribution of Daily New COVID-19 Cases in the United States')
plt.show()
```
