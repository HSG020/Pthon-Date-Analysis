```python
import pandas as pd
import requests
import hashlib
import random
import time

# 百度翻译API的配置信息
appid = 'your_baidu_app_id'  # 替换为你的百度翻译App ID
secretKey = 'your_baidu_secret_key'  # 替换为你的百度翻译密钥
url = 'http://api.fanyi.baidu.com/api/trans/vip/translate'

# 定义翻译函数
def baidu_translate(query, from_lang='en', to_lang='zh'):
    salt = random.randint(32768, 65536)
    sign = appid + query + str(salt) + secretKey
    sign = hashlib.md5(sign.encode()).hexdigest()

    params = {
        'q': query,
        'from': from_lang,
        'to': to_lang,
        'appid': appid,
        'salt': salt,
        'sign': sign,
    }

    try:
        response = requests.get(url, params=params)
        result = response.json()
        return result['trans_result'][0]['dst']
    except Exception as e:
        print(f"Error occurred: {e}")
        return query  # 返回原始文本，防止翻译失败

# 读取CSV文件
df = pd.read_csv('credits.csv')

# 对每个单元格进行翻译
for column in df.columns:
    df[column] = df[column].apply(lambda x: baidu_translate(str(x)))
    time.sleep(1)  # 延时1秒，避免请求过多导致API限制

# 保存翻译后的CSV文件
df.to_csv('translated_file.csv', index=False)

print("翻译完成并保存为translated_file.csv")

```
