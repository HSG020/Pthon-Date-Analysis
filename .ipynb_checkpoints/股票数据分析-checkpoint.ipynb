```python
jz = pd.read_csv("五粮液股票数据.csv", encoding="gbk")
jz.head()
```

```python
jz_cleaned = jz.dropna
```

```python
jz_filled = jz.fillna(0)
```

```python
jz
```

```python
jz["交易日期"] = pd.to_datetime(jz["交易日期"],format="%Y%m%d")
```

```python
filltered_jz = jz[(jz["交易日期"] >= "2018-01-01") & (jz["交易日期"] <= "2018-12-31")]
```

```python
filltered_jz
```

```python
average_open_prince = filltered_jz["open price"].mean()
average_open_prince
```

```python
max_high_price = filltered_jz["high price"].max()
```

```python
print(f"2018年开盘价平均值:{average_open_prince}")
print(f"2018年最高价最大值：{max_high_price}")
```
